```python
# "MANUAL" DE USO

#Para usar este código solamente se necesita darle a 'Run' sección por sección
#para que funcione correctamente

# Importar las librerías a utilizar
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
```

```python
# Leer los datos a analizar
mam = pd.read_csv("Mammal_brains.csv")
print('shape: ', mam.shape)
mam.head() 
```

```python
#Se crea el dataframe y se calcula la varianza
dfmam = pd.DataFrame(mam)
#dfmam.head()
```

```python
#Para saber los tipos de datos en el archivo se utiliza dtypes
dfmam.dtypes
```

```python
dfmam.isnull().sum()
#No hay datos vacíos
```

```python
correlation_matrix = dfmam.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
#Esta es la matriz de correlación en forma de mapa de calor.
```

```python
# Se utiliza displot de la librería seaborn para
# un análisis preliminar
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(dfmam['Brain'], bins=30)
plt.show()
```

```python
# Analisis de relación de características 
# específicas con la variable dependiente
# En este caso, peso del cerebro y masa del animal
sns.lmplot("Body", "Brain", dfmam)
```
